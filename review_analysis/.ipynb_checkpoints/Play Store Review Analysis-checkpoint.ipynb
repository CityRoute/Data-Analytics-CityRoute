```python
from google_play_scraper import Sort, reviews
import json
import spacy
```

## 1 Star Reviews

```python
onestar_result, onestar_continuation_token = reviews(
    'com.dublinbus.wearei3',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    count=300, # defaults to 100
    filter_score_with=1 # defaults to None(means all score)
)
```

```python
for review in onestar_result:
    print(review["content"], end="\n\n")
```

## 5 Star Reviews

```python
fivestar_result, fivestar_continuation_token = reviews(
    'com.dublinbus.wearei3',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    count=300, # defaults to 100
    filter_score_with=5 # defaults to None(means all score)
)
```

```python
for review in fivestar_result:
    print(review["content"], end="\n\n")
```
