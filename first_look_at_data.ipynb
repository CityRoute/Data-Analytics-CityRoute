## Create a sqlite connection to db to find the table info

```python
# Establishing a connection to the database
con = sqlite3.connect('/home/faye/notebooks/rt_leavetimes_DB_2018.db')
```

```python
def sql_fetch(con, query):
    cursorObj = con.cursor()
    cursorObj.execute(query)
    pprint(cursorObj.fetchall())
    return cursorObj.fetchall()
```

```python
# list info about the table in the database
sql_fetch(con, 'select * from sqlite_master;')
```

```python
query_df = query_df.sort_values(by='PROGRNUMBER', ascending=True)
query_df
```

```python
query_df["DIFF_ARR"] = query_df["ACTUALTIME_ARR"] - query_df["PLANNEDTIME_ARR"]
query_df["DIFF_DEP"] = query_df["ACTUALTIME_DEP"] - query_df["PLANNEDTIME_DEP"]
query_df["ACTUAL_DWELL_TIME"]= query_df["ACTUALTIME_ARR"] - query_df["ACTUALTIME_DEP"]
query_df["PLANNED_DWELL_TIME"]= query_df["PLANNEDTIME_ARR"] - query_df["PLANNEDTIME_DEP"]
query_df["DIFF_DWELL"]= query_df["ACTUAL_DWELL_TIME"] - query_df["PLANNED_DWELL_TIME"]
```

```python
fig, ax = plt.subplots()

ax.plot(query_df.ACTUALTIME_ARR, query_df.PROGRNUMBER, label="ACTUALTIME_ARR")
ax.plot(query_df.PLANNEDTIME_ARR , query_df.PROGRNUMBER, label="PLANNEDTIME_ARR")
plt.legend(loc="upper left")
plt.show()
```

```python
def plot_columns(columns, xaxis):
    fig, ax = plt.subplots()
    for column in columns:
        ax.plot(query_df[xaxis], query_df[column], label=column)
    plt.legend(loc="upper left")
    plt.show()
```
