```python
import pandas as pd
import dask.dataframe as dd
import datetime
```

```python
df_trips = pd.read_csv('/home/faye/data/rt_trips_DB_2018.txt', sep=';',error_bad_lines=False)
df_vehicles = pd.read_csv('/home/faye/data/rt_vehicles_DB_2018.txt', sep=';',error_bad_lines=False)
df_leavetimes = dd.read_csv('/home/faye/data/rt_leavetimes_DB_2018.txt', sep=';',error_bad_lines=False)
```

```python
# all columns containing time values (expressed in seconds)
time_columns= ('PLANNEDTIME_ARR', 'PLANNEDTIME_DEP', 'ACTUALTIME_ARR', 'ACTUALTIME_DEP')
```

Converting each time column to a datetime format (more readable):

```python
# note: may take a while as processing 4 columns
for col in time_columns:
    df_trips[col] = pd.to_datetime(df_trips[col], unit='s').dt.strftime("%H:%M:%S")
```
